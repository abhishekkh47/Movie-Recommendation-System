```python
# Now we will use "Collaborative Filtering" to choose "Top-N" from (say) 100 movies ( which the person may liked )
# In the current dataset, the data is not properly arranged
# So we need to arrange it in a better way (in the form of a table [ having movies as columns and UserId as rows])
# This is called Pivot Table
# Pivot Table => Feature Scaling
```

```python
ratings_df.pivot(index='UserId', columns='MoviesId', values='Rating')
```

```python
user_ratings_df = ratings_df.pivot(index='UserId', columns='MoviesId', values='Rating')
```

```python
# Feature Scaling => Applying Normalization (values = 0 to 1)
user_ratings_df/5
```

```python
# but 'NaN' is not acceptable, so convert it to 0
norm_user_rating_df = user_ratings_df.fillna(0) / 5.0
```

```python
type(norm_user_rating_df)
```

```python
# But tensorflow works on 'numpy', not pandas
trX = norm_user_rating_df.values
```

```python
# "Preprocessing" means to convert the data into a form that machine understand, and its done!
user_ratings_df.columns
```

```python
visibleUnits = len(user_ratings_df.columns)
```

```python
visibleUnits
# It means we need to create visible layer of 3706 neurons
# But number of neurons in hidden layer is an hyper parameter, so we randomly check for it
```

```python
hiddenUnits = 20
```

```python
# here  we are not using Keras, we are using Temsorflow
# When we use Keras/Tensorflow, we need to convert int/string into tensor variables
vb = tf.placeholder("float", visibleUnits)
```

```python
vb
```

```python
hb = tf.placeholder("float", [hiddenUnits])
```

```python
W = tf.placeholder("float", [visibleUnits, hiddenUnits])
# This is a kind of 2D-array/ matrix used for calculation and storing the values 
# "W" will maintain the values of bias of each calculation (each neuron of 'movie' with each neuron of 'rating')
# That's why we have created "W" (2D array)
```

```python
W
```

```python
#Phase 1: Input Processing
v0 = tf.placeholder("float", [None, visibleUnits])
_h0 = tf.nn.sigmoid(tf.matmul(v0, W) + hb)
h0 = tf.nn.relu(tf.sign(_h0 - tf.random_uniform(tf.shape(_h0))))
```

```python
#Phase 2: Reconstruction
_v1 = tf.nn.sigmoid(tf.matmul(h0, tf.transpose(W)) + vb) 
v1 = tf.nn.relu(tf.sign(_v1 - tf.random_uniform(tf.shape(_v1))))
h1 = tf.nn.sigmoid(tf.matmul(v1, W) + hb)
```

```python
#Learning rate
alpha = 1.0
#Create the gradients
w_pos_grad = tf.matmul(tf.transpose(v0), h0)
w_neg_grad = tf.matmul(tf.transpose(v1), h1)
#Calculate the Contrastive Divergence to maximize
CD = (w_pos_grad - w_neg_grad) / tf.to_float(tf.shape(v0)[0])
#Create methods to update the weights and biases
update_w = W + alpha * CD
update_vb = vb + alpha * tf.reduce_mean(v0 - v1, 0)
update_hb = hb + alpha * tf.reduce_mean(h0 - h1, 0)
```

```python
err = v0 - v1
err_sum = tf.reduce_mean(err * err)
```

```python
import numpy as np
```

```python
#Current weight
cur_w = np.zeros([visibleUnits, hiddenUnits], np.float32)
#Current visible unit biases
cur_vb = np.zeros([visibleUnits], np.float32)
#Current hidden unit biases
cur_hb = np.zeros([hiddenUnits], np.float32)
#Previous weight
prv_w = np.zeros([visibleUnits, hiddenUnits], np.float32)
#Previous visible unit biases
prv_vb = np.zeros([visibleUnits], np.float32)
#Previous hidden unit biases
prv_hb = np.zeros([hiddenUnits], np.float32)
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
import matplotlib.pyplot as plt
```

```python
epochs = 15
batchsize = 100
errors = []
for i in range(epochs):
    for start, end in zip( range(0, len(trX), batchsize), range(batchsize, len(trX), batchsize)):
        batch = trX[start:end]
        cur_w = sess.run(update_w, feed_dict={v0: batch, W: prv_w, vb: prv_vb, hb: prv_hb})
        cur_vb = sess.run(update_vb, feed_dict={v0: batch, W: prv_w, vb: prv_vb, hb: prv_hb})
        cur_nb = sess.run(update_hb, feed_dict={v0: batch, W: prv_w, vb: prv_vb, hb: prv_hb})
        prv_w = cur_w
        prv_vb = cur_vb
        prv_hb = cur_hb
    errors.append(sess.run(err_sum, feed_dict={v0: trX, W: cur_w, vb: cur_vb, hb: cur_hb}))
    print (errors[-1])
plt.plot(errors)
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.show()
```

```python
#Feeding in the user and reconstructing the input
hh0 = tf.nn.sigmoid(tf.matmul(v0, W) + hb)
vv1 = tf.nn.sigmoid(tf.matmul(hh0, tf.transpose(W)) + vb)
feed = sess.run(hh0, feed_dict={ v0: inputUser, W: prv_w, hb: prv_hb})
rec = sess.run(vv1, feed_dict={ hh0: feed, W: prv_w, vb: prv_vb})
print(rec)
```
