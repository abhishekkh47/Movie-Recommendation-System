# Association Rules

- "Association Rule" is a concept of unsupervised learning
- "Association Rule" have two approahes/algorithms:
 - Apriori method
 - Aclat method

```python
import pandas as pd
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
# in this dataset we don't have any header
```

```python
dataset.info()
```

```python
# We don't have any specific column for one item
# It differs in all transactions
dataset
```

```python
type(dataset)
```

```python
from apyori import apriori
# Library=apyori, Module=apyori, Function=apriori
```

```python
# apriori does not support dataframe
# It supports numpy array
# Here, we have to create a 2-D list (list of records/lists)
# 'iloc' and 'loc' functions are used to retrieve data in dataframe
dataset.loc[0]
```

```python
dataset.iloc[0:1]
```

```python
transactions=[]
for i in range(0, len(dataset)):
    transactions.append( [ str(dataset.values[i,j]) for j in range(0,len(dataset.loc[i])) ] )
```

```python
# to create Association Rules
rules = apriori(transactions=transactions)
```

```python
list(rules)
# Here, we get some "lift<=1", which has no significance
```

```python
rules = apriori(transactions=transactions,
                min_length=2, max_length=2,
                min_support=0.003, min_lift=3,
                min_confidence=0.2)
```

```python
result = list(rules)
```

```python
result
```

```python
# "Eclat" model works similarly but only have gives us "support"
# To apply "Eclat", remove confidence and lift from above code
```

```python
# Clean the data
def inspect(results):
    lhs = [ tuple(result[2][0][0])[0] for result in results ]
    rhs = [ tuple(result[2][0][1])[0] for result in results ]
    supports = [ result[1] for result in results ]
    confidence = [ result[2][0][2] for result in results ]
    lifts = [ result[2][0][3] for result in results ]
    return list( zip( lhs, rhs, supports, confidence, lifts ) )
```

```python
rindf = pd.DataFrame( inspect(result), columns = [ 'Left hand side', 'Right hand side', 'Support', 'Confidence', 'Lift'])
# "inspect()" is used to pass the data(columns) to the function
```

```python
rindf
```

```python
rindf.nlargest(n=10, columns='Lift')
```
